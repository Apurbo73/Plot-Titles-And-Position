```python
import matplotlib.pyplot as plt
import pandas as pd
#Data:
data={
    "Cricket_bat": ["SG", "CA","SS"],
    "MRP": [ 1000,2000,3000],
    "WEIGHT": [1100,1300,1250]

}

# Data Frame
dataFrame= pd.DataFrame(data)
# Plotting line graph:
```

```python
plt.plot(dataFrame["MRP"], dataFrame["WEIGHT"])

#Labels:
plt.xlabel("Bat Prices")
plt.ylabel("Bat Weights")

## Setting Title on the left:
plt.title("Bat Price Depends on Bat's Weight", loc='left')
plt.grid()
```

```python
plt.plot(dataFrame["MRP"], dataFrame["WEIGHT"])

#Labels:
plt.xlabel("Bat Prices")
plt.ylabel("Bat Weights")

## Setting Title on the right:
plt.title("Bat Price Depends on Bat's Weight", loc='right')
plt.grid()
```
